# Night Vision Challenge - Find The Edges

###  The python code is attached along with this file

##### Step 1: Load the libraries

```python
import cv2
from PIL import Image
```

##### Step 2: Get  the image paths and setup output directory

```python
image_path = r'./night_image.jpeg'
final_image = r'./output_night_image.jpeg'
optimized_image = r'./optimized_night_image.jpeg'
output_filename = 'output_night_image.jpeg'
optimizedname = 'optimized_night_image.jpeg'
```

##### Step 3: Display  the basic input file

```python
 Image.open(image_path)
```

##### Step 4: Read  the image_ input file

```python
img = cv2.imread(image_path)
```

##### Step 5: Find the Edge of   the image_ input file

```python
cv2.imwrite(output_filename, cv2.Canny(img,100, 100))
```

##### Step 6: Display the output Image file file

```python
Image.open(final_image)
```

##### Step 7: Optimize the edges

```python
cv2.imwrite(optimizedname, cv2.Canny(img,100, 350))
```

##### Step 8: Display the output optimized Image file

```python
Image.open(optimized_image)
```
